```python
import torch
```

```python
input = torch.ones((1,1,32000))
```

```python
net = torch.nn.Conv1d(1,256,5,1,dilation=8,padding=4)
```

```python
output = net(input)
output.shape
```

```python
layernorm = torch.nn.LayerNorm([256,31976])
```

```python
output2 = layernorm(output)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(output.detach().numpy().reshape(-1))
```
