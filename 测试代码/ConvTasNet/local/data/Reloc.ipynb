```python
import os
import pandas as pd
```

```python
root_dir = '/home/oneran/Wave-Split-CSU2021/测试代码/ConvTasNet/local/data/wav8k/both/'
subdirs = ['dev', 'test', 'train-360']
KEYPATH = 'Libri2Mix'
pwd = '/home/ma-user/work/datasets/Libri2Mix/wav8k/both'
def fun_adjust(df, pwd, flag=False):
    if not flag:
        df[0] = df[0] + '_single'
    end_index = 25565
    for i, word in enumerate(df[1].split('/')):
        if word == 'Libri2Mix':
            end_index = i
    df[1] = pwd + '/' + '/'.join(df[1].split('/')[end_index+3:-1]) + '/' + df[0] + '.wav'
    df[2] = pwd + '/' + '/'.join(df[2].split('/')[end_index+3:-1]) + '/' + df[0] + '.wav'
    df[3] = pwd + '/' + '/'.join(df[3].split('/')[end_index+3:-1]) + '/' + df[0] + '.wav'
    df[4] = pwd + '/' + '/'.join(df[4].split('/')[end_index+3:-1]) + '/' + df[0] + '.wav'
    df[5] = df[5]
    return df


for subdir in subdirs:
    dirpath = root_dir + subdir
    csv = os.listdir(dirpath)[0]
    cur_path = dirpath + '/{}'.format(csv)
    df = pd.read_csv(cur_path)
    for i in range(len(df)):
        df.loc[i] = fun_adjust(df.loc[i], pwd, True)
        df.loc[i+len(df)] = fun_adjust(df.loc[i], pwd)
    df.to_csv(cur_path, index=False)
```
